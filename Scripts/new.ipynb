```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv("../Data/driver_circuit_data/Lewis_Hamilton_Circuit_de_Monaco.csv")  # Replace with actual path

# Set style
sns.set(style="whitegrid", palette="muted")

# Plot
plt.figure(figsize=(12, 6))
plt.plot(data['year'], data['lap_time_seconds'], marker='o', label='Hamilton Lap Time', color='red')
plt.plot(data['year'], data['teammate_time'], marker='o', label='Teammate Lap Time', color='blue')

# Highlight fastest lap (min lap_time_seconds)
min_idx = data['lap_time_seconds'].idxmin()
min_year = data.loc[min_idx, 'year']
min_time = data.loc[min_idx, 'lap_time_seconds']
plt.scatter(min_year, min_time, color='green', s=100, label='Fastest Lap')

# Labels and title
plt.title('Qualifying Lap Times of Lewis Hamilton and teammate at Monaco', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Lap Time (seconds)', fontsize=12)
plt.legend()
plt.xticks(data['year'])  # Ensure all years are shown
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
df = pd.read_csv("../Data/predicted_vs_actual/predicted_vs_actual_2022.csv")  # Replace with your actual CSV path

# Drop rows with missing deviation_percent
df = df.dropna(subset=['deviation_percent'])

# Group by driver and calculate average deviation
driver_deviation = df.groupby('driver')['deviation_percent'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(x=driver_deviation.values, y=driver_deviation.index, palette="coolwarm")
plt.xlabel("Average Prediction Error (%)")
plt.ylabel("Driver")
plt.title("Average Lap Time Prediction Error per Driver (2022)", fontsize=14)
plt.tight_layout()
plt.show()
```
